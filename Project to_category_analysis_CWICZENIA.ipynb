```python
# merge orderlines + products = df ,add 2 extra colum after merge
df = orderlines_df.merge(products_df, how="left", on = "sku")
df['discount_amount']=df['price']-df['unit_price']
df['discount_pect']=(((df['price']-df['unit_price'])*100)/(df['price'])).round(2)
```

Some discounts found negative.Remove this from orders_df and orderlines_df to get cleaned data

```python
# Get order IDs with negative discounts
invalid_orders = df.loc[df['discount_pect'] < 0, 'id_order'].unique()
```

```python
# Remove those orders from orderlines_df
orderlines_df= orderlines_df.loc[~orderlines_df['id_order'].isin(invalid_orders)]
orderlines_df.shape
```

```python
# Remove same orders from orders_df
orders_df = orders_df.loc[~orders_df['order_id'].isin(invalid_orders)]
```

```python
import seaborn as sns
sns.relplot(data=df,
            x="price",
            y="unit_price",
            hue="discount_pect",
            palette="viridis");
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.relplot(data=df,
            x="price",
            y="unit_price",
            hue="discount_pect",
    palette="viridis")

for ax in g.axes.flat:
    ax.axline((0, 0), slope=1, color='blue', linestyle='--', label='Exact Price')

plt.title("MRP vs. Actual Sold Price")
plt.show()
```

```python
# histogram of the discount percentages
sns.displot(data=df,
            x="discount_pect",
            bins=20, kde=True)

plt.title("Frequency of Discount Percentages")
plt.xlabel("Discount %")
plt.ylabel("Number of Products")
plt.show()
```

```python
df_plot = sns.catplot(kind="bar",
            data=df,
            x="discount_pect",
            y='category',
            hue="category",
            palette='viridis',
            height=6,
            aspect=1.5
            )
plt.title("Average Discount Percentage by Category", fontsize=18)
df_plot.set_axis_labels("Category", "Discount_percentage", fontsize=16);
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
#Do higher discounts increase revenue?
plt.figure(figsize=(8,5))
sns.scatterplot(data=all_merge_df, x='discount_pect', y='total_paid', alpha=0.5)
sns.regplot(data=all_merge_df, x='discount_pect', y='total_paid', scatter=False, color='red')

plt.title('Discount % vs Revenue')
plt.xlabel('Discount Percentage')
plt.ylabel('Total Paid')
plt.show()
```

```python
#total revenue by category
plt.figure(figsize=(10,6))
sns.barplot(data=analysis_df, y='category', x='total_paid')

plt.xticks(rotation=90)
plt.title('Total Revenue by Category')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=all_merge_df, y='category', x='discount_pect')

plt.xticks
plt.title('Discount Distribution by Category')
plt.show()
```

```python
# Top 10 most selling Products--identify best performers
top_products = all_merge_df.groupby('name').agg({'product_quantity':'sum'}).reset_index()
top_products = top_products.sort_values(by='product_quantity', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_products, x='product_quantity', y='name')

plt.title('Top 10 Most Selling Products')
plt.xlabel('Quantity Sold')
plt.ylabel('Product')
plt.show()
```

```python
#Sales Over Time (Trend Analysis)--Monthly revenue trend
all_merge_df['date'] = pd.to_datetime(all_merge_df['date'])

time_sales = all_merge_df.groupby(all_merge_df['date'].dt.to_period('M')).agg({'total_paid':'sum'}).reset_index()
time_sales['date'] = time_sales['date'].astype(str)

plt.figure(figsize=(12,6))
sns.lineplot(data=time_sales, x='date', y='total_paid', marker='o')

plt.xticks(rotation=45)
plt.title('Monthly Revenue Trend')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=all_merge_df, x='discount_pect', y='product_quantity', alpha=1)
#sns.regplot(data=all_merge_df, x='discount_pect', y='product_quantity', scatter=False, color='green')

plt.title('Discount % vs Quantity Sold')
plt.xlabel('Discount Percentage')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# moat sold products vs most revenue
product_analysis = all_merge_df.groupby('name').agg({'product_quantity': 'sum','total_paid': 'sum'}).reset_index()

plt.figure(figsize=(10,6))

sns.scatterplot(data=product_analysis,x='product_quantity',y='total_paid')

sns.regplot(data=product_analysis,x='product_quantity',y='total_paid',scatter=False)

plt.title('Most Sold Products vs Revenue')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Total Revenue')
```
